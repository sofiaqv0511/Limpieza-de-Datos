```python
#Importar librerias
import pandas as pd
import numpy as np
```

```python
# 1. Cargar el archivo CSV desde tu ruta local
df = pd.read_csv(r"C:\Documentos\SEXTO SEMESTRE\Ciencia de Datos\archive (1)\Womens Clothing E-Commerce Reviews.csv")

# Mostrar las dimensiones y las primeras filas
print("Dimensiones del dataset:", df.shape)
df.head()
```

```python
# 2. Información general y valores nulos
df.info()

# Ver cuántos valores faltantes hay por columna
df.isnull().sum()
```

```python
# Revisar porcentaje de nulos
(df.isnull().sum() / len(df)) * 100

# Eliminar columnas con más del 50% de nulos
df = df.dropna(thresh=len(df)*0.5, axis=1)

# Rellenar nulos en texto con cadenas vacías
df['Title'] = df['Title'].fillna('Sin título')
df['Review Text'] = df['Review Text'].fillna('Sin reseña')
```

```python
# Contar duplicados
duplicados = df.duplicated().sum()
print(f"Duplicados antes de limpiar: {duplicados}")

# Eliminar duplicados
df = df.drop_duplicates()

print("Duplicados después de limpiar:", df.duplicated().sum())
```

```python
# Revisar estadísticos
df.describe()

# Corregir valores fuera de rango en Rating (1 a 5)
df = df[(df['Rating'] >= 1) & (df['Rating'] <= 5)]

# Eliminar edades fuera del rango típico (por ejemplo, 13–90 años)
df = df[(df['Age'] >= 13) & (df['Age'] <= 90)]
```

```python
import re

def limpiar_texto(texto):
    texto = texto.lower()                          # minúsculas
    texto = re.sub(r'[^a-záéíóúñ\s]', '', texto)  # eliminar símbolos
    texto = re.sub(r'\s+', ' ', texto).strip()     # quitar espacios dobles
    return texto

df['Title'] = df['Title'].apply(limpiar_texto)
df['Review Text'] = df['Review Text'].apply(limpiar_texto)
```

```python
# Revisar valores únicos
print("Division Name:", df['Division Name'].unique())
print("Department Name:", df['Department Name'].unique())

# Convertir a formato uniforme (capitalizado)
df['Division Name'] = df['Division Name'].str.title()
df['Department Name'] = df['Department Name'].str.title()

# Reemplazar errores comunes si los hay
df['Department Name'] = df['Department Name'].replace({
    'Intimates': 'Intimate', 
    'Bottoms ': 'Bottoms'
})
```

```python
print("Dimensiones finales:", df.shape)
print("Valores nulos finales:\n", df.isnull().sum())
df.head()
```
